```python
"""
Created on Thu Mar 26 18:13:05 2020

@author: ayaz
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('dataset.csv')
dataset.head()
```

```python
X = dataset['Head Size(cm^3)'].values
Y = dataset['Brain Weight(grams)'].values
```

### Note that the equation for linear regression is:
#### Y = mX + C
![](formula.jpeg)

```python
numerator = np.sum(np.multiply((X-np.mean(X)),(Y-np.mean(Y))))
denominator = np.sum(np.square((X-np.mean(X))))
m = numerator/denominator
c = np.sum(np.mean(Y)-(m*np.mean(X)))
print(m,c)
```

### Plotting the points

```python
plt.scatter(X,Y, color = "magenta", label = "Scatterplot")
plt.xlabel('Head Size (cm^3)')
plt.ylabel('Brain Weight (grams)')
```

### Plotting the line

```python
x_max = np.max(X)+100;
x_min = np.min(X)-100;

x = np.linspace(x_max, x_min, 1000)
y = m*x+c;

plt.plot(x,y, color = "green", label = "Linear regression line")
plt.xlabel('Head Size (cm^3)')
plt.ylabel('Brain Weight (grams)')
```

### Fitting Linear Regression

```python
x_max = np.max(X)+100;
x_min = np.min(X)-100;

x = np.linspace(x_max, x_min, 1000)
y = m*x+c;

plt.plot(x,y, color = "green", label = "Linear regression line")
plt.scatter(X,Y, color = "magenta", label = "Scatterplot")

plt.xlabel('Head Size (cm^3)')
plt.ylabel('Brain Weight (grams)')

plt.legend()
plt.show()
```
